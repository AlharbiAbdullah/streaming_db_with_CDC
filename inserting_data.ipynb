```python
#import needed libraries
from sqlalchemy import create_engine
import pandas as pd
import time
from dotenv import dotenv_values

#.env values 
config = dotenv_values('.env')


#get password from environmnet var
uid = config['POSTGRES_USER']
pwd = config['POSTGRES_PASSWORD']
server = "localhost"
db = config['POSTGRES_DB']
port = "5432"


engine = create_engine(f'postgresql://{uid}:{pwd}@{server}:{port}/{db}')
```

```python
# Reading from Source table 
df = pd.read_sql('Select * from public.customers', engine)
df.head()
```

```python
for index, row in df.iterrows():
    mod = pd.DataFrame(row.to_frame().T)
    mod.to_sql(f"customers_streaming", engine, if_exists='append', index=False)
    print("Row Inserted " + mod.customer_id.astype(str) + ' ' + mod.contact_name.astype(str).astype(str))
    time.sleep(3)
```
